# PHASE 1: DATA CLEANING & EXPLORATION

## Step 1: Load Data

```python
print("\n" + "="*100)
print("PHASE 1: DATA CLEANING & EXPLORATION")
print("="*100 + "\n")

print("STEP 1: LOADING DATA")
print("-"*100)

df = pd.read_csv('Country-data.csv')
print(f"Dataset Shape: {df.shape}")
print(f"Countries: {df.shape[0]}")
print(f"Features: {df.shape[1]}")
print(f"\nFirst 5 rows:")
print(df.head())

country_names = df['country'].values
df_features = df.iloc[:, 1:]
```

## Step 2: Data Quality Check

```python
print("\n\nSTEP 2: DATA QUALITY CHECK")
print("-"*100)

missing_data = df.isnull().sum()
print(f"Missing values per column:\n{missing_data}")

if missing_data.sum() > 0:
    print(f"\n⚠️  Removing {len(df) - len(df.dropna())} rows with missing values")
    df = df.dropna()
    country_names = df['country'].values
    df_features = df.iloc[:, 1:]

duplicates = df.duplicated().sum()
print(f"Duplicate rows: {duplicates}")

if duplicates > 0:
    df = df.drop_duplicates()
    country_names = df['country'].values
    df_features = df.iloc[:, 1:]

print(f"\n✅ Final dataset: {df.shape[0]} countries × {df.shape[1]-1} features")
```

## Step 3: Descriptive Statistics

```python
print("\n\nSTEP 3: DESCRIPTIVE STATISTICS")
print("-"*100)
print(df_features.describe())
```

## Step 4: Outlier Detection (IQR Method)

```python
print("\n\nSTEP 4: OUTLIER DETECTION (IQR Method)")
print("-"*100)

def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]

outlier_summary = {}
for col in df_features.columns:
    outliers = detect_outliers_iqr(df, col)
    outlier_summary[col] = len(outliers)

print(f"Outliers per column: {outlier_summary}")
print(f"Total outlier instances: {sum(outlier_summary.values())}")
```

## Step 5: Correlation Analysis

```python
print("\n\nSTEP 5: CORRELATION ANALYSIS")
print("-"*100)
correlation_matrix = df_features.corr()
print("\nHighly Correlated Features (|r| > 0.7):")
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        if abs(correlation_matrix.iloc[i, j]) > 0.7:
            print(f"  {correlation_matrix.columns[i]} ↔ {correlation_matrix.columns[j]}: {correlation_matrix.iloc[i, j]:.3f}")
```

## Step 6: Feature Range Analysis

```python
print("\n\nSTEP 6: FEATURE RANGE ANALYSIS")
print("-"*100)
print(f"\n{'Feature':<20} {'Min':<15} {'Max':<15} {'Range':<15}")
print("-"*65)
for col in df_features.columns:
    feature_range = df_features[col].max() - df_features[col].min()
    print(f"{col:<20} {df_features[col].min():>13.2f}  {df_features[col].max():>13.2f}  {feature_range:>13.2f}")

# Save cleaned data
df.to_csv('country_data_cleaned.csv', index=False)
print("\n✅ Saved: country_data_cleaned.csv")
```

# PHASE 2: PCA IMPLEMENTATION

## Step 1: Feature Standardization

```python
print("\n\n" + "="*100)
print("PHASE 2: PCA IMPLEMENTATION")
print("="*100 + "\n")

print("STEP 1: FEATURE STANDARDIZATION")
print("-"*100)

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_features)
df_scaled = pd.DataFrame(df_scaled, columns=df_features.columns)

print("Before Standardization - Mean & StdDev:")
print(f"  Mean: {df_features.mean().mean():.2f}")
print(f"  StdDev: {df_features.std().mean():.2f}")

print("\nAfter Standardization - Mean & StdDev:")
print(f"  Mean: {df_scaled.mean().mean():.6f}")
print(f"  StdDev: {df_scaled.std().mean():.2f}")
print("✅ Standardization complete")
```

## Step 2: Fit Full PCA Model

```python
print("\n\nSTEP 2: FITTING FULL PCA MODEL")
print("-"*100)

n_components = df_scaled.shape[1]
pca_full = PCA(n_components=n_components)
pca_transformed_full = pca_full.fit_transform(df_scaled)

explained_variance = pca_full.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)
eigenvalues = pca_full.explained_variance_

print(f"\nPCA fitted with {n_components} components")
print(f"\n{'Component':<12} {'Eigenvalue':<15} {'Variance %':<15} {'Cumulative %':<15}")
print("-"*60)
for i in range(n_components):
    print(f"PC{i+1:<10} {eigenvalues[i]:>13.4f} {explained_variance[i]*100:>13.2f}% {cumulative_variance[i]*100:>13.2f}%")
```

## Step 3: Determine Optimal Components

```python
print("\n\nSTEP 3: DETERMINING OPTIMAL COMPONENTS")
print("-"*100)

variance_threshold = 0.95
n_components_optimal = np.argmax(cumulative_variance >= variance_threshold) + 1
n_components_optimal = min(n_components_optimal, 5)

print(f"Components needed for 95% variance: {n_components_optimal}")
print(f"Variance explained: {cumulative_variance[n_components_optimal-1]*100:.2f}%")
```

## Step 4: Fit Optimal PCA Model

```python
print("\n\nSTEP 4: FITTING OPTIMAL PCA MODEL")
print("-"*100)

pca = PCA(n_components=n_components_optimal)
pca_transformed = pca.fit_transform(df_scaled)

pca_df = pd.DataFrame(
    pca_transformed,
    columns=[f'PC{i+1}' for i in range(n_components_optimal)]
)
pca_df['country'] = country_names

loadings = pca.components_
loadings_df = pd.DataFrame(
    loadings.T,
    columns=[f'PC{i+1}' for i in range(n_components_optimal)],
    index=df_features.columns
)

print(f"✅ Fitted PCA with {n_components_optimal} optimal components")
print(f"   Dimensionality reduction: {df_scaled.shape[1]} → {n_components_optimal} ({(1-n_components_optimal/df_scaled.shape[1])*100:.1f}% reduction)")
```

## Step 5: Component Interpretation

```python
print("\n\nSTEP 5: COMPONENT INTERPRETATION")
print("-"*100)

for pc_idx in range(n_components_optimal):
    pc_name = f'PC{pc_idx+1}'
    pc_loading = loadings_df[pc_name]
    top_pos = pc_loading.nlargest(3)
    top_neg = pc_loading.nsmallest(3)
    
    print(f"\n{pc_name}: ({explained_variance[pc_idx]*100:.2f}% variance)")
    print(f"  Positive contributors:")
    for feat, val in top_pos.items():
        print(f"    ✓ {feat:20s}: {val:+.3f}")
    print(f"  Negative contributors:")
    for feat, val in top_neg.items():
        print(f"    ✗ {feat:20s}: {val:+.3f}")

# Save PCA results
pca_df.to_csv('pca_transformed_data.csv', index=False)
loadings_df.to_csv('pca_loadings.csv')
print("\n✅ Saved: pca_transformed_data.csv")
print("✅ Saved: pca_loadings.csv")
```

# PHASE 3: CLUSTERING IMPLEMENTATION

## Step 1: Determine Optimal Number of Clusters

```python
print("\n\n" + "="*100)
print("PHASE 3: CLUSTERING IMPLEMENTATION")
print("="*100 + "\n")

print("STEP 1: ELBOW METHOD - FIND OPTIMAL CLUSTERS")
print("-"*100)

# Try different numbers of clusters
inertias = []
silhouette_scores = []
K_range = range(2, 11)

for k in K_range:
    kmeans_temp = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans_temp.fit(df_scaled)
    inertias.append(kmeans_temp.inertia_)
    silhouette_scores.append(silhouette_score(df_scaled, kmeans_temp.labels_))

print(f"\n{'K':<5} {'Inertia':<15} {'Silhouette':<15}")
print("-"*35)
for k, inertia, sil_score in zip(K_range, inertias, silhouette_scores):
    print(f"{k:<5} {inertia:<15.2f} {sil_score:<15.4f}")

print(f"\n✅ For this analysis, we use k=3 (Most Developed, Developed, Least Developed)")
```

```python
# Plot elbow curve
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('Optimal Cluster Selection', fontsize=16, fontweight='bold')

# Inertia plot
ax = axes[0]
ax.plot(K_range, inertias, 'bo-', linewidth=2, markersize=8)
ax.axvline(x=3, color='red', linestyle='--', linewidth=2, label='k=3 (Selected)')
ax.set_xlabel('Number of Clusters (k)', fontsize=12, fontweight='bold')
ax.set_ylabel('Inertia (Within-cluster sum of squares)', fontsize=12, fontweight='bold')
ax.set_title('Elbow Method', fontsize=13, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend(fontsize=11)

# Silhouette plot
ax = axes[1]
ax.plot(K_range, silhouette_scores, 'go-', linewidth=2, markersize=8)
ax.axvline(x=3, color='red', linestyle='--', linewidth=2, label='k=3 (Selected)')
ax.set_xlabel('Number of Clusters (k)', fontsize=12, fontweight='bold')
ax.set_ylabel('Silhouette Score', fontsize=12, fontweight='bold')
ax.set_title('Silhouette Analysis', fontsize=13, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend(fontsize=11)

plt.tight_layout()
plt.savefig('11_elbow_silhouette.png', dpi=300, bbox_inches='tight')
print("✅ Saved: 11_elbow_silhouette.png")
plt.show()
```

```python
print("\n\nSTEP 2: FIT K-MEANS WITH k=3")
print("-"*100)

# Fit K-Means with k=3 for development tiers
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clusters = kmeans.fit_predict(df_scaled)

print(f"\nK-Means fitted with k=3 clusters")
print(f"Cluster distribution: {np.bincount(clusters)}")

# Calculate metrics
silhouette_avg = silhouette_score(df_scaled, clusters)
davies_bouldin = davies_bouldin_score(df_scaled, clusters)
calinski_harabasz = calinski_harabasz_score(df_scaled, clusters)

print(f"\nClustering Quality Metrics:")
print(f"  Silhouette Score: {silhouette_avg:.4f} (range: -1 to 1, higher is better)")
print(f"  Davies-Bouldin Index: {davies_bouldin:.4f} (lower is better)")
print(f"  Calinski-Harabasz Index: {calinski_harabasz:.2f} (higher is better)")
```

## Step 2: Classify Countries into Development Tiers

```python
print("\n\nSTEP 3: CLASSIFY INTO 3 DEVELOPMENT TIERS")
print("-"*100)

# Calculate cluster statistics based on development indicators
cluster_stats = {}
for i in range(3):
    cluster_mask = clusters == i
    cluster_stats[i] = {
        'life_expec': df_features.loc[cluster_mask, 'life_expec'].mean(),
        'gdpp': df_features.loc[cluster_mask, 'gdpp'].mean(),
        'child_mort': df_features.loc[cluster_mask, 'child_mort'].mean(),
        'income': df_features.loc[cluster_mask, 'income'].mean(),
        'count': cluster_mask.sum()
    }

# Rank clusters by development indicators
print("\nCluster Characteristics:")
for cluster_id in range(3):
    stats = cluster_stats[cluster_id]
    print(f"\nCluster {cluster_id}:")
    print(f"  Countries: {stats['count']}")
    print(f"  Avg Life Expectancy: {stats['life_expec']:.2f} years")
    print(f"  Avg GDP per capita: ${stats['gdpp']:,.2f}")
    print(f"  Avg Child Mortality: {stats['child_mort']:.2f} per 1000")
    print(f"  Avg Income Index: {stats['income']:.3f}")
```

```python
# Sort clusters by development level (life expectancy + GDP)
cluster_ranking = sorted(cluster_stats.items(), 
                        key=lambda x: (x[1]['life_expec'], x[1]['gdpp']), 
                        reverse=True)

# Map clusters to development tiers
tier_mapping = {
    cluster_ranking[0][0]: 'Most Developed',
    cluster_ranking[1][0]: 'Developed',
    cluster_ranking[2][0]: 'Least Developed'
}

print("\n" + "-"*100)
print("DEVELOPMENT TIER ASSIGNMENTS:")
print("-"*100)

for rank_idx, (cluster_id, stats) in enumerate(cluster_ranking):
    tier = tier_mapping[cluster_id]
    # Now 'stats' is already the dictionary, not a tuple!
    print(f"Life Expectancy: {stats['life_expec']:.2f} years")
    
    print(f"\n{rank_idx+1}. {tier.upper()} (Cluster {cluster_id})")
    print(f"   Countries: {int(stats['count'])}")
    print(f"   Life Expectancy: {stats['life_expec']:.2f} years")
    print(f"   GDP per Capita: ${stats['gdpp']:,.2f}")
    print(f"   Child Mortality: {stats['child_mort']:.2f} per 1000 births")
    print(f"   Income Index: {stats['income']:.3f}")
```

```python
# Create classification dataframe
tier_classification_df = pd.DataFrame({
    'Country': country_names,
    'Cluster': clusters,
    'Development_Tier': [tier_mapping[c] for c in clusters]
})

# Add all features
for feature in df_features.columns:
    tier_classification_df[feature] = df_features[feature].values

print("\n\nSTEP 4: COUNTRY LISTINGS BY DEVELOPMENT TIER")
print("-"*100)

# Display countries by tier
for tier in ['Most Developed', 'Developed', 'Least Developed']:
    tier_countries = sorted(tier_classification_df[tier_classification_df['Development_Tier'] == tier]['Country'].values)
    print(f"\n{tier.upper()} ({len(tier_countries)} countries):")
    # Print in rows of 4
    for i in range(0, len(tier_countries), 4):
        print(f"  {', '.join(tier_countries[i:i+4])}")
```

```python
# Save classification results
print("\n\nSTEP 5: EXPORT RESULTS")
print("-"*100)

tier_classification_df.to_csv('country_development_tiers.csv', index=False)
print(f"\n✅ Saved: country_development_tiers.csv")
print(f"   Rows: {len(tier_classification_df)}")
print(f"   Columns: {len(tier_classification_df.columns)}")
```
